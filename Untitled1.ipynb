```python
from scipy.interpolate import UnivariateSpline
from scipy.signal import savgol_filter

# 读取数据
df = GMST
start_date = '1992-04'
end_date = '2024-03'
mask = (df.index >= start_date) & (df.index <= end_date)
filtered = df[mask].copy()

# 计算年度平均值（4月-次年3月）
def fiscal_year(date):
    if date.month >= 4:
        return date.year + 1
    else:
        return date.year

annual_means = filtered.resample('MS').mean().groupby(fiscal_year).mean()
years = annual_means.index.values
values = annual_means['GMST'].values

# Enting样条拟合 --------------------------------------------
def enting_spline(years, values, cut_off_period=30):
    """
    Enting样条滤波实现
    参数：
    years : 年度序列（数值型数组）
    values : 对应年度的温度值
    cut_off_period : 截止周期（年）
    """
    # 计算平滑参数（基于截止周期）
    time_span = years[-1] - years[0]
    s_ratio = (cut_off_period / time_span)**2
    
    # 三次样条拟合
    spline = UnivariateSpline(years, values, s=s_ratio*len(values))
    trend = spline(years)
    
    return trend

# 执行拟合（30年截止周期）
trend_30yr = enting_spline(years, values, cut_off_period=30)

# 可视化结果 ------------------------------------------------
plt.figure(figsize=(12, 6))

# 原始年度均值
plt.plot(years, values, 
         color='steelblue', 
         alpha=0.7, 
         label='Annual Mean')

# 拟合趋势线
plt.plot(years, trend_30yr, 
         color='crimson', 
         linewidth=2.5,
         label='Enting Spline (30-year cutoff)')

plt.title('Global Mean Surface Temperature Trend Analysis\n(1983-2024 Annual Averages)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# 输出趋势数据 ----------------------------------------------
result_df = pd.DataFrame({
    'Year': years,
    'Annual_Mean': values,
    'Trend_30yr': trend_30yr
}).set_index('Year')

print(result_df.tail(10))
```
